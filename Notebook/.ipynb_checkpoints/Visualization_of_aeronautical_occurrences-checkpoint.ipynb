## 1. Introduction and Context

Safety in civil aviation is a fundamental pillar for the transportation of passengers and cargo, requiring constant vigilance and continuous process improvement. The Center for Investigation and Prevention of Aeronautical Accidents (CENIPA) provides a rich dataset on aeronautical occurrences within Brazilian territory, which serves as a valuable tool for understanding the causes and patterns behind incidents and accidents.

This project aims to perform an in-depth exploratory data analysis (EDA) of this dataset, transforming raw information into actionable insights that can contribute to the improvement of operational safety.

## 2. Business Problem Definition

A civil aviation regulatory body needs to optimize the allocation of its resources (inspection teams, training programs, prevention campaigns) to reduce the number and severity of aeronautical occurrences. Currently, these actions are distributed generally, but it is believed that a **data-driven approach** can identify key risk areas and make safety initiatives more effective.

The challenge of this analysis is, therefore, to **investigate the historical data of occurrences to provide a clear diagnosis of the main points of concern in Brazilian civil aviation**, answering key questions that will guide the agency's strategic decision-making.

## 3. Questions to be Answered

To address the business problem, this analysis will follow a logical roadmap, seeking to answer the following questions:

#### A. Descriptive Analysis (The Big Picture)
1.  **Historical Trend:** Has the number of occurrences (accidents and incidents) been increasing or decreasing over the years?
2.  **Severity:** What is the proportion of occurrences by classification (accident, serious incident, incident)?
3.  **Geographical Focus:** Which states and cities have the highest concentration of occurrences?
4.  **Seasonality:** Are there specific months or times of the year when events are more frequent?

#### B. Diagnostic Analysis (Investigating the Causes)
5.  **Aircraft Profile:** What types of aircraft (manufacturer, engine type) are most frequently involved in occurrences, especially the most severe ones?
6.  **Type of Operation:** Which flight purpose (e.g., agricultural, instructional, air taxi) shows the highest incidence of records?
7.  **Contributing Factors:** What are the main factors (human, material, operational) that lead to an occurrence? Does the prevalence of these factors change with the severity of the event?
8.  **Phases of Flight:** In which phase of flight (takeoff, landing, cruise) are the risks highest?

#### C. Recommendations and Actions
9.  **Actionable Insights:** Based on the patterns found, what would be the priority recommendations for the regulatory body? (e.g., intensify inspections in a specific segment, focus training campaigns on a particular contributing factor, etc.)

```python
# Define the datasets and their file paths
dataset_files = {
    'aeronave': '../Data/aeronave.csv',
    'fator_contribuinte': '../Data/fator_contribuinte.csv',
    'ocorrencia_tipo': '../Data/ocorrencia_tipo.csv',
    'ocorrencia': '../Data/ocorrencia.csv',
    'recomendacao': '../Data/recomendacao.csv'
}

# Create a function to safely read a CSV file with fallback encoding
def load_csv_safely(filepath, sep=';', encodings=['utf-8', 'latin1']):
    for enc in encodings:
        try:
            return pd.read_csv(filepath, sep=sep, encoding=enc)
        except Exception:
            continue
    raise ValueError(f"Could not read file {filepath} with encodings {encodings}")

# Load all datasets into a dictionary
datasets = {}
for name, path in dataset_files.items():
    if os.path.exists(path):
        datasets[name] = load_csv_safely(path)
        print(f"Loaded: {name} ({datasets[name].shape[0]} rows, {datasets[name].shape[1]} columns)")
    else:
        print(f"File not found: {path}")
```

We will fill in the missing data in each dataset with values that make sense and do not harm the analysis.

### Renaming the Columns
Since the codigo_ocorrencia columns represent the same thing, the unique key of an occurrence, we will rename them to make it easier.

```python
# Treating ocorencia
# Removing the codigo_ocorrencia4 column which is not useful now
columns_to_drop = ['divulgacao_dia_publicacao', 'codigo_ocorrencia4']
ocorrencia.drop(columns=columns_to_drop, inplace=True)

# Filling the missing categorical and numeric values in this dataset
ocorrencia['ocorrencia_latitude'].fillna(value=np.nan, inplace=True)
ocorrencia['ocorrencia_longitude'].fillna(value=np.nan, inplace=True)
ocorrencia['divulgacao_relatorio_numero'].fillna('Não divulgado', inplace=True)
ocorrencia['ocorrencia_hora'].fillna('00:00', inplace=True)
ocorrencia['investigacao_status'].fillna('Desconhecido', inplace=True)
```

```python
# Treating ocorrencia_tipo
ocorrencia_tipo.fillna('Nao informado', inplace=True)
```

```python
# Treating recomendacao
for col in recomendacao.columns:
    if col != 'recomendacao_dia_feedback':
        recomendacao[col].fillna('Nao informado', inplace=True)
```

```python
# Treating aeronave
# Dividing the dataset columns by categorical and numeric types
object_cols = aeronave.select_dtypes(include='object').columns
# Creating a list  os numeric columns that need treatment, excluding the identification column "codigo_ocorrencia"
cols_to_impute = aeronave.select_dtypes(include=np.number).columns.drop('codigo_ocorrencia2')

# Filling categorical and numeric null values
aeronave[object_cols] = aeronave[object_cols].fillna('Desconhecido')

for col in cols_to_impute:
    median = aeronave[col].median()
    aeronave[col] = aeronave[col].fillna(median)
    print(f"Null values in column {col} filled with the median: {median}")
```

```python
# Standardizing the key column to 'codigo_ocorrencia' in all tables to facilitate joins and analysis integrity.

ocorrencia_tipo.rename(columns={'codigo_ocorrencia1': 'codigo_ocorrencia'}, inplace=True)
aeronave.rename(columns={'codigo_ocorrencia2': 'codigo_ocorrencia'}, inplace=True)
fator_contribuinte.rename(columns={'codigo_ocorrencia3': 'codigo_ocorrencia'}, inplace=True)
recomendacao.rename(columns={'codigo_ocorrencia4': 'codigo_ocorrencia'}, inplace=True)

# Now we can discard the others codigo_ocorrencia in ocorrencia
ocorrencia = ocorrencia.drop(columns=['codigo_ocorrencia1', 'codigo_ocorrencia2', 'codigo_ocorrencia3'])
```

## Answering Question A1: Historical Trend

```python
# Ensure the 'ocorrencia_dia' column is in datetime format
# The format DD/MM/YYYY requires dayfirst=True

ocorrencia['ocorrencia_dia'] = pd.to_datetime(ocorrencia['ocorrencia_dia'], dayfirst=True, errors='coerce')

# Create a simple function to execute the queries
pysqldf = lambda q: sqldf(q, globals())

# Extract the year into a new column
ocorrencia['ano'] = ocorrencia['ocorrencia_dia'].dt.year

#Aggregate data: Count occurences per year
query_top_years = """
    SELECT 
        ano, COUNT(codigo_ocorrencia) as codigo_ocorrencia_cont
    FROM 
        ocorrencia
    GROUP BY 
        ano
    ORDER BY 
        ano
"""
ocorrencias_por_ano = pysqldf(query_top_years)

# Visualization
plt.figure(figsize=(14, 7))

# Create the line plot
sns.lineplot(data=ocorrencias_por_ano, x='ano', y='codigo_ocorrencia_cont', marker='o', color='navy')

# Adding title ans labels for clarity
plt.title('Historical Trend of Aeronautical Occurrences in Brazil', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Occurences', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# Diplay the plot
plt.show()
```

## Answering Question A3: Geographical Focus

```python
# We select the state (ocorrencia_uf), count all occurrences for that state,
# group the results by state, and order them to see the ones with the most occurrences.
# We limit to the top 10 for a cleaner visualization.

query_top_states = """
    SELECT 
        ocorrencia_uf, COUNT(*) as total_occurences
    FROM
        ocorrencia_filtrada
    WHERE
        ocorrencia_uf IS NOT NULL AND ocorrencia_uf != '***'
    GROUP BY
        ocorrencia_uf
    ORDER BY
        total_occurences DESC
    LIMIT 10;"""

# Execute the query using the function we created
top_10_states = pysqldf(query_top_states)

# Display the results (Wich is a pandas dataframe)
print("Top 10 States by Number of occurences (Result from SQL query)")
print(top_10_states)

# Visualize the results from the SQL query
plt.figure(figsize=(12, 6))
sns.barplot(data = top_10_states, x='ocorrencia_uf', y ='total_occurences', palette='viridis')

plt.title('Top 10 States with Most Aeronautical Occurrences', fontsize=16, fontweight='bold')
plt.xlabel('State (UF)', fontsize=12)
plt.ylabel('Number of Occurrences', fontsize=12)
plt.show()
```

```python
# First, let's identify the Top 5 states from our filtered data
top_5_states_list = ocorrencia_filtrada['ocorrencia_uf'].value_counts().nlargest(5).index.tolist()
print("Top 5 States identified for analysis:", top_5_states_list)

# Merge the occurrence and aircraft dataframes to link state with operation type
# We use an 'inner' join to ensure we only work with records that have complete information in both tables.
df_merged = pd.merge(ocorrencia_filtrada, aeronave, on='codigo_ocorrencia', how='inner')

# Filter this new merged dataframe to keep only the records from our top 5 states
df_top5 = df_merged[df_merged['ocorrencia_uf'].isin(top_5_states_list)]

# Aggregate the data: Count each 'aeronave_tipo_operacao' within each state
# The .unstack() method pivots the data, making it perfect for a stacked bar chart
op_counts = df_top5.groupby(['ocorrencia_uf', 'aeronave_tipo_operacao']).size().unstack(fill_value=0)

# Clean up for visualization: Consolidate less frequent categories into an 'Other' category
# This prevents the chart from having too many tiny, unreadable segments.
# First, identify the top 5 most frequent operation types overall in these states.
top_operations = op_counts.sum().nlargest(5).index.tolist()
print("\nTop 5 Operation Types being analyzed:", top_operations)

# Create an 'Other' column by summing up all columns that are NOT in our top list
op_counts['Other'] = op_counts.drop(columns=top_operations).sum(axis=1)

# Now, we keep only the main operation types and the new 'Other' column
op_counts_final = op_counts[top_operations + ['Other']]

# Lastly, reorder the states in the dataframe to match the original ranking (most occurrences first)
op_counts_final = op_counts_final.reindex(top_5_states_list)

# Plot the stacked bar chart
op_counts_final.plot(kind='bar', 
                     stacked=True, 
                     figsize=(14, 8),
                     colormap='viridis')

# Adding titles and labels for clarity
plt.title('Breakdown of Occurrence by Operation Type in Top 5 States', fontsize=16, fontweight='bold')
plt.xlabel('State (UF)', fontsize=12)
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Type of Operation', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()

plt.show()
```

### A3: Geographical Distribution and Operational Profile by State

**Primary Observation: Absolute Numbers**
An initial analysis of occurrences by state clearly shows that **São Paulo (SP)** has a significantly higher absolute number of events than any other state, followed by major economic hubs like Rio de Janeiro (RJ) and Minas Gerais (MG).

**Context and Hypothesis: Volume vs. Risk**
This concentration is strongly hypothesized to be a direct result of higher air traffic volume in these states, which host the country's busiest airports. To conduct a fair comparison of relative risk, the data would need to be **normalized** by the total number of flight operations (e.g., takeoffs). As this operational data is not part of the current dataset, we cannot conclude that flying in São Paulo is inherently more dangerous; only that it is significantly busier.

**Deeper Dive: Operational Profile by State**
To understand the nature of these occurrences beyond just the raw count, a breakdown by `Type of Operation` for the top 5 states was created. The stacked bar chart reveals distinct operational risk profiles:

* **Major Hubs (SP & RJ):** In São Paulo and Rio de Janeiro, occurrences are dominated by **'REGULAR'** (commercial) and **'PRIVADA'** (private) flights. Rio de Janeiro also shows a significant share of **'TÁXI AÉREO'** operations. This profile is consistent with their roles as the nation's primary centers for commercial and executive aviation.

* **States with Diverse Aviation (MG, PR, RS):** In contrast, Minas Gerais, Paraná, and Rio Grande do Sul show a more varied mix. While 'PRIVADA' operations remain a major factor, there is a much larger proportional contribution from **'AGRÍCOLA'** (agricultural) and **'INSTRUÇÃO'** (instructional) flights compared to SP and RJ.

**Strategic Implication for the Regulatory Body:**
This detailed breakdown confirms that a "one-size-fits-all" safety strategy would be inefficient. The data strongly suggests that a regulatory body should create targeted interventions based on the unique risk profile of each state. For example:
* Safety initiatives in **São Paulo and Rio de Janeiro** could focus on improving air traffic control and ground procedures related to high-density commercial and executive flights.
* In **Paraná, Minas Gerais, and Rio Grande do Sul**, resources could be more effectively directed towards specialized training programs and enhanced safety protocols for the agricultural and flight instruction sectors.

### A4: Seasonality Analysis

The next step is to investigate whether occurrences follow a seasonal pattern. The initial hypothesis was that events would be more frequent during peak travel seasons in Brazil (December/January and July) due to higher air traffic volume.

**1. Aggregate Monthly Distribution (Bar Chart Analysis)**
The bar chart, which aggregates all occurrences by month across the entire period, reveals a subtle but interesting trend.

* **Observation:** Contrary to the initial hypothesis of dramatic spikes, the distribution is relatively stable, with monthly totals generally ranging between 700 and 900 occurrences. While the end-of-year months (Sep-Dec) show a slight increase, there isn't an overwhelming peak that corresponds with the massive increase in holiday travel.

* **Interpretation:** This relative stability could be interpreted in two ways:
    1.  **Effective Mitigation:** It's plausible that airports and airlines implement heightened safety protocols and allocate more resources during peak seasons, effectively mitigating the risk and preventing a proportional rise in occurrences.
    2.  **Data Lag:** Alternatively, there could be lags in reporting that smooth out the monthly totals over time.

**2. Year-over-Year Consistency (Heatmap Analysis)**
To verify if this pattern is consistent, a heatmap was generated to show the distribution of occurrences by both year and month. This provides a much more granular view.

* **Observation 1 - Distributed Peaks:** The heatmap shows that no single month is consistently the highest-risk across all years. The peak month varies annually (e.g., March in 2007, September in 2010, January in 2014), indicating that while a general seasonal trend exists, it is also influenced by year-specific factors.

### B7: Analysis of Contributing Factors

This analysis investigates the root causes of occurrences by examining the main contributing factors, as categorized by CENIPA. The analysis is twofold: first, identifying the most common factors overall, and second, determining if their prevalence changes with the severity of the event.

**1. Overall Factor Distribution (Based on the Simple Bar Chart)**
The initial analysis shows that **'FATOR OPERACIONAL' (Operational Factor)** and **'FATOR HUMANO' (Human Factor)** are the dominant categories, each accounting for a very large portion of the total contributing factors cited in investigations. 'FATOR MATERIAL' (Material Factor) is cited far less frequently.

* **Human Factor Insights:** The high number of human-factor contributions is a significant area of concern. It prompts critical questions for a regulatory body: Are these errors linked to pilot inexperience (new pilots), complacency (very experienced pilots), fatigue from long duty periods, or other systemic issues within training and operational procedures? While this dataset does not specify the root cause, it clearly flags the human element as a primary area for safety improvement initiatives.

* **Operational Factor Insights:** Operational factors, which include external elements like adverse weather, runway conditions, and bird strikes, are also highly prevalent. As inferred from previous analyses, these events are likely common across all aviation segments but may have a higher impact on the general aviation sector, which is dominated by piston-engine aircraft that are more susceptible to such conditions.

**2. The Critical Shift: How Factor Prevalence Changes with Severity (Based on the Stacked Bar Chart)**
The most significant finding of this analysis emerges from the 100% stacked bar chart. It reveals a dramatic and crucial shift in the nature of contributing factors as the severity of an event increases:

* **In 'Incidents':** Minor events are predominantly driven by **'Operational Factors'**. This makes intuitive sense, as many external issues (like a bird strike or a minor system glitch) will be classified as an incident. The 'Material Factor' also shows its highest prevalence here, suggesting that component failures are often caught and resolved before they escalate.

* **In 'Accidents':** The profile completely inverts for the most severe outcomes. The proportion of the **'FATOR HUMANO' (Human Factor)** grows, becoming more significant in accidents. As the human factor's share increases, the relative contributions of operational and material factors decrease significantly.

**Strategic Implication for the Regulatory Body:**
This analysis provides a clear, data-driven directive for any safety agency. To reduce the overall *frequency* of all events, addressing common operational factors is important.

However, to reduce the *severity* of events and prevent the most catastrophic outcomes **accidents** the primary and most critical focus **must** be on mitigating the **Human Factor**. This confirms that the highest-leverage strategy for saving lives and preventing substantial damage is to invest in areas such as:
* Advanced pilot training, especially in decision-making under pressure (Crew Resource Management - CRM).
* Robust policies for managing pilot fatigue.
* Improved human-machine interfaces in cockpits to reduce the likelihood of error.

```python
# --- 1. DATA PREPARATION FOR EACH DASHBOARD PANEL ---

# Data for Chart 1 (B5): Get Top 10 Manufacturers involved in ACCIDENTS.
# This uses the pre-filtered dataframe containing only accident data.
df_accidents_only = df_b5_analysis[df_b5_analysis['ocorrencia_classificacao'] == 'ACIDENTE']
top_10_fabricantes_accidents = df_accidents_only[~df_accidents_only['aeronave_fabricante'].isin(['***', 'UNKNOWN'])]['aeronave_fabricante'].value_counts().nlargest(10).reset_index()
top_10_fabricantes_accidents.columns = ['Manufacturer', 'Count']

# Data for Chart 2 (B5): Get Engine Type distribution for ACCIDENTS.
engine_types_accidents = df_accidents_only[~df_accidents_only['aeronave_motor_tipo'].isin(['***', 'UNKNOWN'])]['aeronave_motor_tipo'].value_counts().reset_index()
engine_types_accidents.columns = ['Engine Type', 'Count']

# Data for Chart 3 (B7): Create a normalized crosstab of Severity vs. Factor Area.
# This will be used to create the 100% stacked bar chart.
df_b7_analysis = pd.merge(ocorrencia_filtrada, fator_contribuinte, on='codigo_ocorrencia', how='inner')
severity_factor_crosstab = pd.crosstab(df_b7_analysis['ocorrencia_classificacao'], df_b7_analysis['fator_area'])
severity_factor_perc = severity_factor_crosstab.div(severity_factor_crosstab.sum(axis=1), axis=0) * 100
severity_factor_perc = severity_factor_perc.reindex(['INCIDENTE', 'INCIDENTE GRAVE', 'ACIDENTE'])

# Data for Chart 4 (B8): Get Top 10 most frequent flight phases.
top_10_phases = df_b5_analysis[~df_b5_analysis['aeronave_fase_operacao'].isin(['***', 'UNKNOWN', 'INDETERMINADA'])]['aeronave_fase_operacao'].value_counts().nlargest(10).reset_index()
top_10_phases.columns = ['Phase', 'Count']


# --- 2. INITIALIZE THE DASHBOARD LAYOUT ---

# Create a 2x2 grid for the charts, setting spacing and subplot titles.
fig = make_subplots(
    rows=2, cols=2,
    specs=[[{}, {}],
           [{}, {}]],
    subplot_titles=("Top 10 Manufacturers (in ACCIDENTS)", 
                    "Engine Types (in ACCIDENTS)",
                    "Top 10 Flight Phases by Occurrences",
                    "Prevalence of Factors by Severity"),
    horizontal_spacing=0.12, # Adjusts the space between columns
    vertical_spacing=0.2     # Adjusts the space between rows
)

# --- 3. POPULATE THE GRID WITH CHARTS ---

# Panel [1,1]: Top Manufacturers in Accidents
fig.add_trace(go.Bar(y=top_10_fabricantes_accidents['Manufacturer'], x=top_10_fabricantes_accidents['Count'],
                     orientation='h', name='Manufacturers', marker_color='#d62728', showlegend=False), row=1, col=1)

# Panel [1,2]: Engine Types in Accidents
fig.add_trace(go.Bar(x=engine_types_accidents['Engine Type'], y=engine_types_accidents['Count'],
                     name='Engine Types', marker_color='#8c564b', showlegend=False), row=1, col=2)

# Panel [2,1]: Top Flight Phases
fig.add_trace(go.Bar(y=top_10_phases['Phase'], x=top_10_phases['Count'],
                     orientation='h', name='Phases', showlegend=False), row=2, col=1)

# Panel [2,2]: Factors by Severity (100% Stacked Bar)
# Iterate through each factor area (column) to create a stacked bar for each severity level.
for factor in severity_factor_perc.columns:
    fig.add_trace(go.Bar(x=severity_factor_perc.index, y=severity_factor_perc[factor],
                         name=factor), row=2, col=2)


# --- 4. FINALIZE LAYOUT AND STYLING ---

# Update the overall layout of the dashboard figure
fig.update_layout(
    title_text="<b>Dashboard: Diagnostic Analysis of Occurrences (Block B)</b>",
    title_x=0.5, # Center the main title
    height=900,
    template='plotly_white',
    barmode='stack', 
    showlegend=True,
    legend_title_text='Factor Area',
    margin=dict(t=100)
)

# Update the axis titles and ordering for each subplot
fig.update_yaxes(categoryorder='total ascending', row=1, col=1)
fig.update_yaxes(categoryorder='total ascending', row=2, col=1) 
fig.update_layout(barmode='stack', bargap=0.1) 

# Update titles for each axis for clarity
fig.update_xaxes(title_text="Number of Accidents", row=1, col=1)
fig.update_yaxes(title_text="Manufacturer", row=1, col=1)
fig.update_xaxes(title_text="Engine Type", row=1, col=2)
fig.update_yaxes(title_text="Number of Accidents", row=1, col=2)
fig.update_xaxes(title_text="Number of Occurrences", row=2, col=1)
fig.update_yaxes(title_text="Phase of Flight", row=2, col=1)
fig.update_xaxes(title_text="Severity", row=2, col=2)
fig.update_yaxes(title_text="Proportion of Factors (%)", row=2, col=2)

# Display the final dashboard
fig.show()
```

# C9. Conclusions and Actionable Insights

This final section synthesizes the key findings from the descriptive (Block A) and diagnostic (Block B) analyses to provide a set of prioritized, data-driven recommendations for the regulatory body.

### Executive Summary

The analysis reveals that a "one-size-fits-all" approach to aviation safety in Brazil is inefficient. While commercial aviation is remarkably safe, the majority of risk is concentrated in **General Aviation**. Occurrences in this sector, particularly involving **piston-engine aircraft**, are the most frequent.

The most critical finding is the dynamic nature of contributing factors. While Operational Factors are the most common cause across all events, the **proportion of the Human Factor with event severity**, becoming a critical contributor in the accidents. Therefore, to achieve the greatest impact on reducing the *severity* of occurrences, resources must be strategically focused on mitigating Human Factor risks, especially within General Aviation during the takeoff and landing phases.

### Key Findings

The following data-supported findings form the basis for our recommendations:

1.  **The Human Factor's Prevalence Increases with Severity:** This is the most important diagnostic finding. While 'Operational Factors' are the most frequent cause in minor incidents, the **'Human Factor's' contribution grows significantly in severe events**, becoming nearly as prevalent as operational issues in 'Accidents'. This indicates that the human element is a primary driver of event escalation from a minor incident to a major accident.

2.  **General Aviation is the Epicenter of Risk:** The **'Private'**, **'Instructional'**, and **'Agricultural'** sectors collectively account for the largest share of all occurrences. Aircraft used in these operations, predominantly powered by **piston engines**, are more frequently involved in severe accidents compared to the commercial jet fleet.

3.  **Risk Profiles are Regional and Operational:** The *type* of risk is not uniform across the country. Different states show different dominant operational profiles (e.g., commercial hubs vs. agricultural states), requiring tailored safety approaches.

4.  **Takeoff and Landing are the Most Critical Phases:** A vast majority of all occurrences are concentrated in the initial and final phases of a flight. The cruise phase, despite its duration, is statistically the safest.

### Priority Recommendations for the Regulatory Body

Based on these findings, the following actions are recommended, in order of priority, to maximize their impact on aviation safety:

**1. (Highest Priority) Launch a Targeted Program to Mitigate the Human Factor in General Aviation:**
This is the single most important action to reduce the number of severe accidents.
* **Action:** Review and enhance the training curricula for private pilot licenses (PPL), focusing on Crew Resource Management (CRM for single pilot operations), risk assessment, and decision-making under pressure.
* **Action:** Develop mandatory simulator training modules that replicate the most common accident scenarios related to human error during takeoff and landing.
* **Action:** Initiate studies and campaigns on pilot fatigue and mental well-being within the general aviation community.

**2. Implement Segment-Specific Safety Oversight:**
Move away from a uniform inspection model and tailor oversight to the specific risks of each operational type.
* **Action:** Increase ramp inspections and create specialized safety protocols for agricultural aviation in states where this practice is prevalent (e.g., Paraná, Minas Gerais).
* **Action:** Reinforce runway safety protocols and air traffic procedures at high-volume airports (e.g., São Paulo) to reduce the frequency of minor incidents.

**3. Deepen the Analysis of Piston-Engine Aircraft Accidents:**
* **Action:** Commission a focused study on the most common models of piston-engine aircraft involved in accidents. The goal is to identify if there are recurring patterns of material failure that could be addressed through Airworthiness Directives or if the human-machine interface presents specific challenges.

**4. Address Data Quality and Reporting:**
* **Action:** Formally document the change in data reporting methodology that occurred in 2023 to ensure future time-series analyses are accurate. Strive for consistent data collection standards to prevent future structural breaks that could be misinterpreted as sudden changes in safety levels.
